# Merging Datasets
Use Pandas Merges to create a combined dataset from `clean_08.csv` and `clean_18.csv`.

```python
# load datasets
import pandas as pd

df_08 = pd.read_csv('clean_2008.csv')
df_18 = pd.read_csv('clean_2018.csv')
```

### Create combined dataset

```python
df_08
```

```python
df_08.drop(columns=['Unnamed: 0', 'record_year'], inplace=True)
df_18.drop(columns=['Unnamed: 0', 'record_year'], inplace=True)
```

```python
# rename 2008 columns
df_08.rename(columns=lambda x: x[:10] + "_2008", inplace=True)
```

```python
# view to check names
df_08.head()
```

```python
# merge datasets
df_combined = df_08.merge(df_18, left_on='model_2008', right_on='model', how='inner')
```

```python
#this is a custom function to open each dataset
def open_fuel_year(fuel_year):
    
    df = pd.read_csv('./all_alpha_'+str(fuel_year)+'.csv', low_memory=False)

    df['record_year'] = pd.Series(dtype='str')
    new_cols_order = ['record_year'] + df.columns[0:18].to_list()
    
    df = df[new_cols_order]
    for year in years:
        if year == fuel_year:
            df['record_year'] = int(fuel_year)
            print(new_cols_order)    
        
    return df
```

```python
fuel_df = pd.DataFrame()

years = [2008,2018]

#join the datasets
for fuel_year in years:
    fuel_df = pd.concat([fuel_df, open_fuel_year(fuel_year)])
    
    print(len(fuel_df))
fuel_df.reset_index(drop=True, inplace=True)
```

```python
#this is a custom function to open each dataset
def open_fuel_year(fuel_year):
    
    df = pd.read_csv('./data_'+str(fuel_year)+'_v2.csv', low_memory=False)

    df['record_year'] = pd.Series(dtype='str')
    new_cols_order = ['record_year'] + df.columns[0:18].to_list()
    
    df = df[new_cols_order]
    for year in years:
        if year == fuel_year:
            df['record_year'] = int(fuel_year)
            print(new_cols_order)    
        
    return df
```

```python
clean_df = pd.DataFrame()

years = [2008,2018]

#join the datasets
for fuel_year in years:
    clean_df = pd.concat([clean_df, open_fuel_year(fuel_year)])
    
    print(len(clean_df))
clean_df.reset_index(drop=True, inplace=True)
```

```python
v08 = clean_df.query('record_year == 2008')
v18 = clean_df.query('record_year == 2018')
```

```python
# First, let's get all the hybrids in 2008
hb_08 = v08[v08['fuel'].str.contains('/')]
hb_08
```

```python
# hybrids in 2018
hb_18 =v18[v18['fuel'].str.contains('/')]
hb_18
```

```python
df1 = hb_08.copy()  # data on first fuel type of each hybrid vehicle
df2 = hb_08.copy()  # data on second fuel type of each hybrid vehicle
df3 = hb_18.copy()
df4 = hb_18.copy()
```

```python
df1
```

```python
# columns to split by "/"
split_columns = ['fuel', 'air_pollution_score', 'city_mpg', 'hwy_mpg', 'cmb_mpg', 'greenhouse_gas_score']
split_columns2 = ['fuel', 'city_mpg', 'hwy_mpg', 'cmb_mpg']

# apply split function to each column of each 08 dataframe copy
for c in split_columns:
    df1[c] = df1[c].apply(lambda x: x.split("/")[0])
    df2[c] = df2[c].apply(lambda x: x.split("/")[1])
    
# apply split function to each column of each 18 dataframe copy
for c in split_columns2:
    df3[c] = df3[c].apply(lambda x: x.split("/")[0])
    df4[c] = df4[c].apply(lambda x: x.split("/")[1])
```

```python
# combine dataframes to add to the original dataframes
new_rows = pd.concat([df1,df2])
new_rows2 = pd.concat([df3,df4])

# now we have separate rows for each fuel type of each vehicle!
new_rows
```
